## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Loading & Inspecting Data

#### Allocation df loading & Cleanning

```python
## Read each single sheet into df

df1 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=0)
df2 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=1)
df3 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=2)
df4 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=3)
df5 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=4)
df6 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=5)
df7 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=6)
df8 = pd.read_excel("Case 3 - Demand Managenet.xlsx", sheet_name=7)

## Dropping Full Empty Rows from the DFs

df1.dropna(axis=0,how='all',inplace=True)
df2.dropna(axis=0,how='all',inplace=True)
df3.dropna(axis=0,how='all',inplace=True)
df4.dropna(axis=0,how='all',inplace=True)
df5.dropna(axis=0,how='all',inplace=True)
df6.dropna(axis=0,how='all',inplace=True)
df7.dropna(axis=0,how='all',inplace=True)
df8.dropna(axis=0,how='all',inplace=True)

#Adding the Country column to each df before combinning them

df1['Country'] = 'Central_USA'
df2['Country'] = 'Lat_Pa'
df3['Country'] = 'AF'
df4['Country'] = 'MEA'
df5['Country'] = 'MER'
df6['Country'] = 'CSE'
df7['Country'] = 'Subsaharan'
df8['Country'] = 'F_east'
df8
```

```python
Main_df=pd.concat([df1, df2,df3,df4,df5,df6,df7,df8], axis=0)  # Combinning all the sheets int one data frame
Main_df
```

```python
Main_df.columns = Main_df.iloc[0] ##sets the first row as column names
Main_df = Main_df[1:]
Main_df.rename(columns = {'Central_USA':'Country'}, inplace = True)

## renaming country column

Main_df.head(5)
```

```python
Main_df2=Main_df.melt(id_vars = ('DIM','Task','APPROVAL','Country'), var_name = 'Month-Year', value_name = 'Hours')
## unpivot month columns to one column ( Month-Year)
Main_df2
```

```python
Main_df2['Emp_ID'] = np.where(Main_df2['DIM'].str.contains('^EMP.*'), Main_df2['DIM'],'')
Main_df2['Emp_ID']=Main_df2['Emp_ID'].replace(r'^\s*$', np.nan, regex=True)
Main_df2['Emp_ID']=Main_df2['Emp_ID'].fillna(method='ffill')
Main_df2['DIM']=np.where(Main_df2['DIM'].str.contains('^EMP.*'), '',Main_df2['DIM'])
Main_df2['DIM']=Main_df2['DIM'].replace(r'^\s*$', np.nan, regex=True)

## seperating the emp id info to seperate column and replace it in the Dim Column

Main_df2
```

```python
Main_df3=Main_df2.dropna(subset=['DIM'])  
## drop the rows with empty employee id in dim column after moving the emp id to seperate column

Main_df3['Hours'] = (
    pd.to_numeric(Main_df3['Hours'],
                  errors='coerce')
      .fillna(0)
    )
## converting the hours column to float instead of object and replace non numeric values with zeros
Main_df3.info()
```

```python
Main_df3.head()
```

```python
Main_df3['Task'] = Main_df3['Task'].fillna('NaN')   ## filling null values in task and approval columns
Main_df3['APPROVAL'] = Main_df3['APPROVAL'].fillna('NaN')
Main_df3.info()
```

```python
Main_df4=pd.pivot_table(Main_df3, values = 'Hours', index=['Country','Emp_ID','Month-Year','APPROVAL','Task'], columns = 'DIM').reset_index()

Main_df4  ## Pivot the dimension column 
```

```python
Main_df4[['Employee_ID1','Employee_ID2']]=Main_df4.Emp_ID.str.split("_", expand = True)
Main_df4[['Month','Year']]=Main_df4['Month-Year'].str.split(" ", expand = True)

Main_df4=Main_df4.drop(['Employee_ID1', 'Emp_ID'], axis=1)
# Renaming Columns
Main_df4.rename(columns = {'Employee_ID2':'Employee_ID'}, inplace = True)
Main_df4.head()
```

```python
Main_df4.info()
```

#### Employee df loading & Cleanning

```python
## Reading the employee sheets
Employee1 = pd.read_excel('Employee_Table.xlsx', sheet_name='Employee Dept')
Employee2 = pd.read_excel('Employee_Table.xlsx', sheet_name='Employee HR Cost')

#setting the first row as column names
Employee1.columns = Employee1.iloc[0]
Employee2.columns = Employee2.iloc[0]

Employee1=Employee1.iloc[1:, :]      ##removes the first row
Employee2=Employee2.iloc[1:, :] 

Employee1.info()
```

```python
#Merging the two tables

Employee = Employee1.merge(Employee2, left_on='Employee', right_on='Employee')

# Dropping the first null Columns
del Employee[Employee.columns[0]]
del Employee[Employee.columns[4]]

Employee.info()
```

```python
Employee.head()
```

```python
# Renaming Columns
Employee.rename(columns = {'Employee':'Employee_ID','LEVEL-01':'Department','LEVEL-02':'Grade','LEVEL-03':'Level'}, inplace = True)
Employee.head()
```

```python
#Convert to integer
Employee['Cost']=pd.to_numeric(Employee['Cost'])
Employee.info()
```

```python
#Cleanning and renaming Columns

Employee[['Level6','level7']]=Employee.Level.str.split("_", expand = True)
Employee[['Grade1','Grade2']]=Employee.Grade.str.split("_", expand = True)
Employee[['Department1','Department2']]=Employee.Department.str.split("_", expand = True)
Employee[['Employee_ID1','Employee_ID2']]=Employee.Employee_ID.str.split("_", expand = True)

Employee=Employee.drop(['Employee_ID1', 'Employee_ID','Level6', 'Grade1','Department1','Level','Grade','Department'], axis=1)
#'Employee_ID2':'Employee_ID',
# Renaming Columns
Employee.rename(columns = {'Employee_ID2':'Employee_ID','Department2':'Department','Grade2':'Grade','level7':'Level'}, inplace = True)
Employee.head()
```

### Exporting Data

```python
# Exporting the two DFs to CSVs

Employee.to_csv("Employees.csv")
Main_df4.to_csv("Allocation.csv")
```
